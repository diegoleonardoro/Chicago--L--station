```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="Landmarks")

landmarks_zipcode_Nominatim = []

for i in range(len(IndividualLandmarks)): 
    data = geolocator.reverse([IndividualLandmarks['LATITUDE'][i],IndividualLandmarks['LONGITUDE'][i]]) 
    try:
        zipcode = data.raw['address']['postcode']
        landmarks_zipcode_Nominatim.append(zipcode)  
    except KeyError:
        landmarks_zipcode_Nominatim.append('Nan')
```
